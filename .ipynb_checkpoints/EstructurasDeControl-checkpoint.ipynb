1.1 Selección (sentencias condicionales)
- Selección de una de varias alternativas en base a alguna condición.
- Indentación para estructurar el código
- Importante el símbolo ‘:’

```python
# # Asigno el bvalo uno a la variable x
x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
```

- Las condiciones son expresiones que se evalúan a True o False.
- Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.

```python
x = 3
y = [1, 2, 3]

print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
```

```python
 x = -3

if x > 0:
    print('Valor positivo')
elif x == 0:
    print('Valor nulo')
else:
    print('Valor negativo')
```

```python
val = 120
if val > 0:
    if val < 100:
        print('Valor positivo')
    else:
        print('Valor muy positivo')
else:
    print('Valor negativo')
```

```python
a = -6

match a:
    case _ if a > 0:
        print('positive')
    case _ if a == 0:
        print('zero')
    case _ if a < 0:
        print('negative')
```

**3 Iteración (bucles)**
- Repetición de un bloque de código.
- La terminación del bucle depende del tipo de bucle.
- Dos tipos: while y for. <br>
**Bucle ‘while’**
- Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta que una condición evalue a False).
- Si la condición evalua a False desde el principio, el bloque de código nunca se ejecuta.
- Cuidado con los bucles infinitos.

Formato general:<br>
while test:        # Mientras se cumple la condición>br>
    statements     # Instrucciones a ejecutar

```python
# # Ejemplo: Mostrar los primeros 3 objetos de una lista

indice = 0
numeros = [9, 4, 7, 1, 2]

while indice < 3:
    print(numeros[indice])
    indice =indice+1
```

```python
# Ejemplo: contar y mostrar los números inferiores a 10.

numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0

while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1
    
print('Contador:', contador)
print('Indice:', indice)
```

```python
 nombre = 'Pablo'

while nombre: # Mientras 'nombre' no sea vacío
    print(nombre)
    nombre = nombre[1:]
```

```python
 # #bucle infinito
i = 0

while i < 10:
    print(i)
    i += 1
```

**Bucle ‘for’**
- Permite recorrer los items de una sequencia o un objeto iterable.
- Funciona en strings, listas, tuplas, etc.<br>
Formato general:<br>
for item in objeto: # Asigna los items del objeto a la variable item en cada iteración<br>
statements # Instrucciones a ejecutar

```python
peliculas = ['Matrix', 'The purge', 'Avatar', 'Star Wars']

for pelicula in peliculas:
    print(pelicula)
```

- También se usa para iterar un número preestablecido de veces (counted loops):

```python
for i in range(10):
    print(i)
```

- range es útil en combinación con len porque permite acceder a los elementos de una secuencia por posición.

```python
nombre = 'Pablo'

for i in range(len(nombre)):
    print(nombre[i])
```

```python
for letra in nombre:
    print(letra)
```

-Iteración de tuplas:

```python
tuplas = [(1, 2,4), (3, 4), (5, 6)]
for a, b,c in tuplas:
    print(a, b,c)
```

- Iteración de diccionarios. Se iteran las claves:

```python
diccionario = {'a': 1, 'b': 2, 'c': 3}

for key in diccionario:
    print(f"{key} => {diccionario[key]}")
```

-  Para iterar los pares (clave-valor) o únicamente los valores, se deben usar los métodos items y values.

```python
for key, value in diccionario.items():
    print(f"{key} => {value}")
```

```python
for value in diccionario.values():
    print(value)
```

- La variable item en la cabecera del for puede ser cualquier expresión que sea válida como parte izquierda de una asignación convencional

```python
for a, b, c in [(1, 2, 3), (4, 5, 6)]:
    print(a, b, c)
```

**3.1 Las sentencias break, continue y else** <br>
**Break y continue**
- Solo tienen sentido dentro de bucles.
- Break permite terminar el bucle por completo.
- Continue permite saltar a la siguiente iteración, continuando con el bucle.
- Pueden aparecer en cualquier parte de un bucle, pero normalmente aparecen dentro de sentencias condicionales (if).


```python
# # Ejemplo: break

rating_to_find = 4.2
movie_ratings = [4.3, 2.5, 1.7, 4.2, 3.8, 3.3, 4.5]

for rating in movie_ratings:
    print(rating)
    if rating == rating_to_find:
        print("Found")
        break
print(rating)
```

```python
for i in range(10):
    if i % 2 == 0:
    continue        # Si el número es par, salta a la siguiente iteración␣
↪
    print(f'Odd number {i}')
```

- Observa como la sentencia continue te puede ayudar a reducir el número de niveles de anidamiento.
- Sin continue el anterior ejemplo sería:

```python
for i in range(10):
    if i % 2 != 0:
        print('Numero impar:', end=' ')
        print(i)
```

```python
for i in range(5):
    for a in range(2):
        print(f"{i} - {a}")
        if i == 3 and a == 0:
            break
```

**Else**<br>
- Los bucles pueden tener una sensencia else.
- Resulta poco intuitiva para muchos programadores porque esta sintaxis no existe en otros lenguajes.
- Se ejecuta cuando el bucle termina con normalidad; es decir, cuando no termina a causa de un break.

```python
lista = [60,60,30,60]
element_to_find = 60

for element in lista:
    if element == element_to_find:
        print("found")
        break
else:
    print("not found")
```

```python
# with flag
lista = [10,30,50,40]
element_to_find = 30
found = False

for element in lista:
    if element == element_to_find:
        found = True
        break
if found:
    print("dato encontrado")
else:
    print("not found")
```

**4 List Comprehensions**<br>
- Permiten construir listas a través de la ejecución repetida (sentencia for) de una expressión para cada item de un objeto iterable.
- Van entre ‘[’ y ’]’. Esto es indicativo de que estamos construyendo una lista.<br>
Sintaxis:<br>
[<expression> for <item> in <iterable>]<br>
Ejemplo: repetir los carácteres de un string

```python
[char*3 for char in "Enrique"]
```

```python
lista = []

for char in 'Enrique':
    lista.append(char)
    
print(lista)
```

Comúnmente, el item de la sentencia for aparecerá en la expresión principal, pero eso no es obligatorio.

```python
[2 for _ in 'Enrique']
```

**4.0.1 Versión extendida**<br>
Se puede especificar un filtro (sentencia if) para obtener únicamente los elementos que cumplan cierta condición.<br>
Sintaxis:<br>
[<expression> for <item> in <iterable> if <condition>]<br>
Ejemplo: obtener números pares:

```python
[y for y in range(9) if y % 2 == 0]
```

-Las list comprehension no son realmente requeridas, ya que siempre podemos escribir un bucle equivalente.


```python
pares = []
for x in range(9):
    if x % 2 == 0:
        pares.append(x)
        
print(pares)
```

**4.0.2 Versión completa**<br>
La sentencia if de una list comprehension también puede contener una expresión alternativa.<br>
Sintaxis:<br>
[<expression_1> if <condition> else <expression_2> for <item> in <iterable>]<br>
Ejemplo: poner a cero los números pares.

```python
[0 if x % 2 == 0 else x for x in range(9)]
```

```python
[x**2 if x > 2 else x for x in range(-4,5) if x > 0]
```

```python
lista = []

for x in range(-4,5):
    if x > 0:
        if x > 2:
            lista.append(x**2)
        else:
            lista.append(x)
            
print(lista)
```

```python
[x**2 if x > 0 else 100 if x==0 else x for x in range(-4,5)]
```

```python
lista = []

for x in range(-4,5):
    if x > 0:
        lista.append(x**2)
    elif x == 0:
        lista.append(100)
    else:
        lista.append(x)
        
print(lista)
```

**4.0.3 Anidamiento**<br>
Las list comprehensions soportan anidamiento en sus expresiones.<br>
Ejemplo: bucle anidado para obtener una lista de tuplas que combinan los elementos de dos listas dadas

```python
lista_1 = [1, 2, 3]
lista_2 = [4, 5]

[(x, y) for x in lista_1 for y in lista_2]
```

**4.0.4 Pros y contras**
- Principales ventajas de las comprehensions en comparación a un bucle convencional:<br>
    – Expresión compacta y legible, si estás familiarizado con la sintaxis.<br>
    – Mejor rendimiento.
- Desventaja: no escalan bien. Una list comprehension se puede convertir rápidamente en una expresión difícil de entender.

**4.0.5 Otros tipos de comprehensions**<br>
**Dictionary comprehensions**<br>
Usando ‘{’ y ‘}’ como delimitadores y una expresión ‘clave : valor’, se obtiene un diccionario en lugar de una lista.<br>
Ejemplo: diccionario donde cada valor es el cuadrado de la clave.

```python
d = {x : x*x for x in range(10)}

print(d)
print(type(d))
```

```python
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]
shopping = {k:v for k,v in zip(items,price)}
print(shopping)
```

```python
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]
for k in zip(items, price):
    print(type(k))
    print(k)
```

```python
#Ejercicio 4
palabras = ["computadora", "avion", "astronomia", "python"]

encontrada = False

for palabra in palabras:
    if palabra.startswith("a") and len(palabra) > 9:
        print("Palabra encontrada:", palabra)
        encontrada = True
        break

if not encontrada:
    print("No se encontró ninguna palabra que cumpla la condición")
```

```python
#Ejercicio 6
dic = {'a': 4.3, 'b': 1, 'c': 7.8, 'd': -5}

clave_max = None
valor_max = float("-inf")

for clave, valor in dic.items():
    if valor > valor_max:
        valor_max = valor
        clave_max = clave

print("La clave con el valor más alto es:", clave_max)
```
